```python
import html
import re
import httpx
```

```python
def get_summary(q: str) -> str:
    """Fetches a short summary from Wikipedia based on the search term."""
    response = httpx.get(
        "https://en.wikipedia.org/w/api.php",
        params={
            "action": "query",
            "list": "search",
            "srsearch": q,
            "format": "json",
        },
        timeout=20.0,
    )
    response.raise_for_status()

    raw_snippet = response.json()["query"]["search"][0]["snippet"]
    clean_snippet = re.sub(r"<[^>]+>", "", raw_snippet)
    return html.unescape(clean_snippet)


query = "elon musk"
summary = get_summary(query)
print(f"Query: {query}")
print(summary if summary else "No results found.")
```

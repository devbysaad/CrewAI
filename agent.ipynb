```python
import html
import re
import httpx
```

```python
def get_summary(q: str) -> str:
    """Fetch a short Wikipedia search snippet for `q`.

    Returns an empty string when there are no results.
    """
    r = httpx.get(
        "https://en.wikipedia.org/w/api.php",
        params={
            "action": "query",
            "list": "search",
            "srsearch": q,
            "format": "json",
        },
        timeout=20.0,
    )
    r.raise_for_status()

    data = r.json()
    results = (data.get("query") or {}).get("search") or []
    if not results:
        return ""

    raw_snippet = results[0].get("snippet", "")
    clean_snippet = re.sub(r"<[^>]+>", "", raw_snippet)
    return html.unescape(clean_snippet)
```

```python
query = "elon musk"
summary = get_summary(query)
print(f"Query: {query}")
print(summary if summary else "No results found.")
```
